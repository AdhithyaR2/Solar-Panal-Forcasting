```python
from re import findall
from os.path import join, basename
from glob import glob

def get_image_name(file_path: str):
    file_name = basename(file_path)
    return findall('[0-9]+', file_name)[-1]
   
images_paths = glob(join(DATA_PATH, 'C:\\Users\dhara\Downloads\dataset_1/images/*.jpg'))
images_paths = sorted(images_paths, key=get_image_name)

annotations_paths = glob(join(DATA_PATH, 'C:\\Users\dhara\Downloads\dataset_1/annotations/*.json'))
annotations_paths = sorted(annotations_paths, key=get_image_name)
```

```python
import pandas as pd
import json

def get_annotation_data(annotation_path: str):
    with open(annotation_path, 'r', encoding='utf-8') as annotation_file:
        annotation_data = json.load(annotation_file)
    return pd.DataFrame(annotation_data['instances'])

get_annotation_data(annotations_paths[0])
```

```python
from skimage import color, draw
from skimage.measure import label, regionprops
import numpy as np
import flyr
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

def get_mask_from_annotation_data(annotation_data, shape=(512, 640)):
    mask = np.zeros(shape, dtype='int16')
    for index, corner_points in enumerate(annotation_data['corners'], start=1):
        polygon = np.array(
            [[coords['y'], coords['x']] for coords in corner_points]
        )
        instance_mask = draw.polygon2mask(shape, polygon)
        mask[instance_mask] = index
    return mask

# Attach image and annotation paths
annotation_path, image_path = next(zip(annotations_paths, images_paths))

# Get FlyrThermogram class object with all thermal data
thermogram = flyr.unpack(image_path)

# Get mask from annotation file
annotation_data = get_annotation_data(annotation_path)
mask = get_mask_from_annotation_data(annotation_data)

masked_image = color.label2rgb(mask, thermogram.render(palette='grayscale'))

# Showing image and mask extracted
_, ax = plt.subplots(1, 3, figsize=(10,5))    

im = ax[0].imshow(thermogram.celsius, cmap='inferno')
cax = make_axes_locatable(ax[0]).append_axes("right", size="5%", pad=0.05)
plt.colorbar(im, cax=cax, values=np.unique(thermogram.celsius), label='Temperature (C)')    
ax[0].set_title('Thermography image')
ax[0].set_axis_off()

ax[1].imshow(mask)
ax[1].set_title('Label mask')
ax[1].set_axis_off()

ax[2].imshow(masked_image)
ax[2].set_title('Masked image')
ax[2].set_axis_off()

plt.tight_layout()
plt.show()
```

```python
def get_center_from_annotation_data(annotation_data):
    return np.array([
        [coords['x'], coords['y']] for coords in annotation_data['center']
    ])

annotation_path, image_path = next(zip(annotations_paths, images_paths))

thermogram = flyr.unpack(image_path)

annotation_data = get_annotation_data(annotation_path)
centers = get_center_from_annotation_data(annotation_data)

plt.imshow(thermogram.render(palette='grayscale'))
plt.scatter(centers[:, 0], centers[:, 1])
plt.axis('off')
plt.show()
```

```python
def get_metadata(image_path):
    thermogram = flyr.unpack(image_path)
    return thermogram.metadata, thermogram.camera_metadata

metadata = np.array(list(map(get_metadata, images_paths)))
thermal_df = pd.DataFrame(list(metadata[:, 0]))
thermal_df
```

```python
def get_defect_module_label(index):    
    if index in [1, 3, 4, 5]:
        return 1
    elif index == 2:
        return 13
    elif index == 6:
        return 40
    return 18

cell_defect_image_path = images_paths[30:33] + images_paths[57:59] + images_paths[102:104]
cell_defect_annotations_path = annotations_paths[30:33] + annotations_paths[57:59] + annotations_paths[102:104]

generator = zip(cell_defect_image_path, cell_defect_annotations_path)
for index, (image_path, annotation_path) in enumerate(generator, start=1):
    label_mask = get_mask_from_annotation_data(
        get_annotation_data(annotation_path)
    )
    
    defect_label = get_defect_module_label(index)
    defect_binary_mask = np.where(label_mask == defect_label, True, False)
    
    thermal_defect = flyr.unpack(image_path).render_pil(
        mask=defect_binary_mask,
        unit='celsius',
        palette='turbo'
    )
    
    plt.subplot(3, 3, index)
    plt.imshow(thermal_defect)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from exif import Image

def get_camera_metadata(cm):
    return {
        'camera_maker': cm.make,
        'camera_model': cm.model,
        'date_time': cm.date_time,
        'latitude': cm.gps_latitude,
        'longitude': cm.gps_longitude,
        'altitude': cm.gps_altitude,
        #'focal_length': cm.focal_length
    }

def get_extra_camera_metadata(image_path):
    with open(image_path, 'rb') as image_file:
        img = Image(image_file)
    return {
        'subject_area': img.subject_area,
        'focal_lenght': img.focal_length
    }

camera_df = pd.DataFrame([get_camera_metadata(cm) for cm in metadata[:, 1]])
camera_df = camera_df.assign(altitude=camera_df.altitude.abs())
extra_camera_df = pd.DataFrame(list(map(get_extra_camera_metadata, images_paths)))

camera_df = pd.concat([camera_df, extra_camera_df], axis=1)
camera_df
```

```python
def approximate_distance(altitude: float, alpha: float=20, beta: float=30) -> float:
    """
    Calculates an approximation of the distance
    between PV modules fixed in a position and
    the drone during thermal inspection.

    Parameters
    ----------
    alpha : float or int
        Tilt angle of PV modules.
    beta : float or int
        Drone's angle in relation to the glass surface
        of PV modules during thermal inspection.
    altitude : float or int
        Drone's altitude relative to a flat terrain.

    Returns
    -------
    float
        Distance between the drone and PV modules.
    """
    return abs(altitude) / (np.sin(90 - alpha - beta))

(
    camera_df.altitude
             .apply(approximate_distance)
             .plot(kind='kde', title='Distances from drone to PV modules')
);
```
